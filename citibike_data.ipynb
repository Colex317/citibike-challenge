```python
# Import dependency
import pandas as pd
```

```python
# Read in the January 2020 csv files
jan_cb_trip1 = pd.read_csv('Resources/202001-citibike-tripdata_1.csv')
jan_cb_trip2 = pd.read_csv('Resources/202001-citibike-tripdata_2.csv')

# Read in the June 2020 csv files
june_cb_trip1 = pd.read_csv('Resources/202006-citibike-tripdata_1.csv')
june_cb_trip2 = pd.read_csv('Resources/202006-citibike-tripdata_2.csv')
```

# View and perform a union on the January and June of 2020 citibike trip datasets

### Column details:
Trip Duration (seconds)

Start Time and Date

Stop Time and Date

Start Station Name

End Station Name

Station ID

Station Lat/Long

Bike ID

User Type (Customer = 24-hour pass or 3-day pass user; Subscriber = Annual Member)

Gender (Zero=unknown; 1=male; 2=female)

Year of Birth

```python
# Get a brief summary of the June trip two (2) dataset
june_cb_trip2.info()
```

```python
# Perform a union of ALL the January and June cititrip datasets along the rows axis
merged_df = pd.concat([jan_cb_trip1,jan_cb_trip2, june_cb_trip1, june_cb_trip2], ignore_index=True)
display(merged_df.head())
display(merged_df.tail())
```

```python
# Get a brief summary of the combined January and June 2020 citibike trip dataset
merged_df.info()
```

```python
# Replacing the 0/1/2 values in the gender column to unknown, male and female
columns_to_replace = ['gender']

# Replace values in specified columns
for column in columns_to_replace:
    merged_df[column] = merged_df[column].replace({0: 'unknown', 1: 'male', 2: 'female'})

merged_df.head()
```

# Export the combined January dataset as csv

```python
# Export the combined January dataset as CSV file
merged_df.to_csv('Merged_CitiBike_Data/merged_trips.csv', index=False)
```
