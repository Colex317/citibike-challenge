```python
# Import dependency
import pandas as pd
```

```python
# Read in the January 2020 csv files
jan_cb_trip1 = pd.read_csv('Resources/202001-citibike-tripdata_1.csv')
jan_cb_trip2 = pd.read_csv('Resources/202001-citibike-tripdata_2.csv')
```

# View and perform a union on the January citibike trip datasets

### Column details:
Trip Duration (seconds)

Start Time and Date

Stop Time and Date

Start Station Name

End Station Name

Station ID

Station Lat/Long

Bike ID

User Type (Customer = 24-hour pass or 3-day pass user; Subscriber = Annual Member)

Gender (Zero=unknown; 1=male; 2=female)

Year of Birth

```python
# View the January trip one (1) dataset
jan_cb_trip1.head()
```

```python
# Get a brief summary of the January trip one (1) dataset
jan_cb_trip1.info()
```

```python
# View the January trip two (2) dataset
jan_cb_trip2.head()
```

```python
# Get a brief summary of the January trip two (2) dataset
jan_cb_trip2.info()
```

```python
# Perform a union of the two January trip datasets along the rows axis
jan_citibike_trip_df = pd.concat([jan_cb_trip1,jan_cb_trip2], ignore_index=True)
jan_citibike_trip_df.head()
```

```python
# Get a brief summary of the combined January trip dataset
jan_citibike_trip_df.info()
```

```python
# Check for null values in the dataframe
null_values = jan_citibike_trip_df.isnull().sum()

# Display columns with null values and their counts
print("Columns with null values and their counts:")
null_values[null_values > 0]
```

# Export the combined January dataset as csv

```python
# Export the combined January dataset as CSV file
jan_citibike_trip_df.to_csv('Data/jan_citibike_trips.csv', index=False)
```
